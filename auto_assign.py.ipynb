```python
# plot a cell net with shape 12 * 12
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.ticker import NullFormatter, FixedLocator


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(50, 25))

tick_spacing=12
ax1.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax1.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax1.set_xlim([225, 0])
ax1.set_ylim([225, 0])

# refine ax location and direction
ax1.xaxis.set_ticks_position('top')
ax1.invert_xaxis()
ax1.grid()


tick_spacing=24
ax2.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax2.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax2.set_xlim([120, 0])
ax2.set_ylim([120, 0])

# refine ax location and direction
ax2.xaxis.set_ticks_position('top')
ax2.invert_xaxis()
ax2.grid()

# plot point
#gt1 = [(18, 18), (24, 36), (48, 36), (48, 24), (36, 12)]
#gt2 = [((24, 24), (24, 48), (48, 64), (48, 48), (48, 24))]
gt1_x = [18, 24, 48, 48, 36]
gt1_y = [18, 36, 36, 24, 12]
ax1.scatter(gt1_x, gt1_y, s=100, c='g')
gt1_x.append(gt1_x[0])
gt1_y.append(gt1_y[0])
ax1.plot(gt1_x, gt1_y, c='g')

gt2_x = [24, 24, 48, 48, 48]
gt2_y = [24, 48, 64, 48, 24]
gt2_x.append(gt2_x[0])
gt2_y.append(gt2_y[0])
ax2.scatter(gt2_x, gt2_y, s=100, c='r')
ax2.plot(gt2_x, gt2_y, c='r')

plt.show()
```
